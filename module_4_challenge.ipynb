## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df[['unit_price','qty','subtotal']].head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def calc_shipping_price(x):
    if x > 50:
        return 7
    else:
        return 10

df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'] * df['total_weight'].apply(calc_shipping_price)
df[['unit_price','unit_weight','qty','total_weight','shipping_price']].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['total_price'] = df['subtotal'] + df['shipping_price'] + ((df['subtotal'] + df['shipping_price']) * .0925).round(2)
df[['subtotal','shipping_price','total_price']].head()

```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['total_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df[['unit_cost','qty','shipping_price','total_cost']].head()

```

```python
# Create a column for the profit of
# each line using line cost and line price
df['profit'] = df['total_price'] - df['total_cost']
df[['total_price','total_cost','profit']].head()

```

```python
# Sort the updated data by "Total Profit" from highest to lowest
def convert_to_millions(x):
    return x / 1000000


sorted_top_5 = top_5_summary_df.sort_values('profit', ascending=False)

sorted_top_5 = sorted_top_5.rename(columns={'client_id': 'Client ID', 'qty': 'Units Purchased', 'shipping_price': 'Shipping Cost', 'total_price': 'Total Revenue', 'total_cost': 'Total Cost', 'profit': 'Total Profit'})

formatted_top_5 = sorted_top_5[['Client ID', 'Units Purchased', 'Shipping Cost', 'Total Revenue', 'Total Profit']]

#formatted_top_5.columns

formatted_top_5["Shipping Cost"] = formatted_top_5.loc[:,'Shipping Cost'].apply(convert_to_millions).map('${:,.2f}'.format) + 'M'
formatted_top_5['Total Revenue'] = formatted_top_5['Total Revenue'].apply(convert_to_millions).map('${:,.2f}'.format) + 'M'
formatted_top_5['Total Profit'] = formatted_top_5['Total Profit'].apply(convert_to_millions).map('${:,.2f}'.format) + 'M'

formatted_top_5
```
